# Trees Data

```python
import pandas as pd

trees = pd.read_csv("trees_data.csv")
trees.head()
```

```python
num_rows = len(trees)
print(f'There are {num_rows} trees in all of pittsburgh.')
```

```python
trees_counts = trees['neighborhood'].value_counts().sort_values(ascending=False).head(25)
print(trees_counts)
```

```python
trees_counts.plot.bar()
```
